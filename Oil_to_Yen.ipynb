```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df_OilYen = pd.read_csv('http://www.charleswebbmba.com/data/oil-yen.csv')
```

```
df_OilYen['Date'] = pd.to_datetime(df_OilYen['Date'])
```

```
df_OilYen.set_index(df_OilYen['Date'], inplace=True) 
df_OilYen.plot('Date','Yen')
df_OilYen.plot('Date','Oil')
```

```


plt.plot( 'Oil', 'Yen', data=df_OilYen, linestyle='none', marker='o')
plt.xlabel('Oil Price')
plt.ylabel('Yen Price')
plt.title('Price of Yen vs Price of Oil')
plt.show()
```
